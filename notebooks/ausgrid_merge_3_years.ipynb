```python
import sys
import os
# Get the parent directory
parent_folder = os.path.dirname(os.getcwd())
# add the parent directory to the Python path so that the scripts can be imported
sys.path.append(parent_folder)
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pvlib
from modules.utils import read_column_list_from_config
```

```python
fname_2010_2011 = '../data/ausgrid/ausgrid_2010_2011.csv'
fname_2011_2012 = '../data/ausgrid/ausgrid_2011_2012.csv'
fname_2012_2013 = '../data/ausgrid/ausgrid_2012_2013.csv'
df_2010_2011 = pd.read_csv(
    fname_2010_2011, index_col=0, parse_dates=['datetime'])
df_2011_2012 = pd.read_csv(
    fname_2011_2012, index_col=0, parse_dates=['datetime'])
df_2012_2013 = pd.read_csv(
    fname_2012_2013, index_col=0, parse_dates=['datetime'])
```

## Merge the dataframes



```python
df_merged = pd.concat([df_2010_2011, df_2011_2012, df_2012_2013])
# The result has all columns from both dataframes (union of columns)
# Missing columns are filled with NaN values
```

```python
# print the shape of the dataframes
print(df_2010_2011.shape)
print(df_2011_2012.shape)
print(df_2012_2013.shape)
# print the shape of the merged dataframe
print(df_merged.shape)
```

## Find the difference of columns between the three dataframes

```python
# find the difference of columns between the three dataframes

# Get column names for each dataframe
cols_2010_2011 = set(df_2010_2011.columns)
cols_2011_2012 = set(df_2011_2012.columns)
cols_2012_2013 = set(df_2012_2013.columns)

print("Column counts:")
print(f"2010-2011: {len(cols_2010_2011)} columns")
print(f"2011-2012: {len(cols_2011_2012)} columns")
print(f"2012-2013: {len(cols_2012_2013)} columns")
print()

# Find columns that are in 2010-2011 but not in others
only_in_2010_2011 = cols_2010_2011 - cols_2011_2012 - cols_2012_2013
print(f"Columns only in 2010-2011: {len(only_in_2010_2011)}")
if only_in_2010_2011:
    print(sorted(only_in_2010_2011))
print()

# Find columns that are in 2011-2012 but not in others
only_in_2011_2012 = cols_2011_2012 - cols_2010_2011 - cols_2012_2013
print(f"Columns only in 2011-2012: {len(only_in_2011_2012)}")
if only_in_2011_2012:
    print(sorted(only_in_2011_2012))
print()

# Find columns that are in 2012-2013 but not in others
only_in_2012_2013 = cols_2012_2013 - cols_2010_2011 - cols_2011_2012
print(f"Columns only in 2012-2013: {len(only_in_2012_2013)}")
if only_in_2012_2013:
    print(sorted(only_in_2012_2013))
print()

# Find common columns across all three dataframes
common_cols = cols_2010_2011 & cols_2011_2012 & cols_2012_2013
print(f"Common columns across all three: {len(common_cols)}")
print()

# Find columns that are missing in 2012-2013 compared to others
missing_in_2012_2013 = (cols_2010_2011 | cols_2011_2012) - cols_2012_2013
print(f"Columns missing in 2012-2013: {len(missing_in_2012_2013)}")
if missing_in_2012_2013:
    print(sorted(missing_in_2012_2013))
```
